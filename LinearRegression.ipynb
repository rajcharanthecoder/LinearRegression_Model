```python
#linear Regression
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataSet = pd.DataFrame(df.data)
dataSet.columns = df.feature_names
(dataSet.head())
```

```python
x =  np.round(df.target,1)
dataSet['SunLight'] = x.astype(int)
```

```python
dataSet.head()
```

```python
#dividing the datSets Into independent features

X = dataSet.iloc[:,:-1] #independent Feature
Y = dataSet.iloc[:,-1] #dependent Feature

X.head()
```

```python
Y.head()
```

```python
#Linear Regression Model
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
meanSquareError = cross_val_score(lin_reg,X,Y,scoring = 'neg_mean_squared_error',cv = 5)
print(meanSquareError)
mean_mse = np.mean(meanSquareError)
print(mean_mse)
```
